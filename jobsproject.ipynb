```python
import numpy as np
from numpy import *
import re
import pandas as pd
import os
import glob
from pandas import *

#This path contains many csvs of jobs data scraped from multiple job boards using import.io
path = 'Desktop/jobcsvs'

#This finds all the CSVs to append to my new dataframe.
filelist = glob.glob(path + "/*.csv") 

print filelist

# The following code concatenates multiple CSVs with unordered columns (each CSV has a different ordering to the columns)
jobs = pd.concat([pd.read_csv(fl, index_col=[0,1]) for fl in filelist])
jobs.head()
```

```python
#Here I delete one unnecessary column
del jobs['url/_text']
jobs.head()
```

```python
#I isolate the one column I want to clean
salarylist=jobs['salary']

#I split this column into 
salarylist=(salarylist.astype(str).str.split())
# salarylist=salarylist.to_string()

salarylist.head()
```

```python
# List items starting with £ using Regex and replace commas in the strings

def parseNumbers(salary_string):
    return [int(item.replace(',','')) for item in re.findall('£([\d,]+)',salary_string)]

#testing if this works
testcases = ['£23,000 to £100,000','£34,000','£10000', 'depending on benefits']
for testcase in testcases:
    print testcase,parseNumbers(testcase)
```

```python
# If there are two such items found, call the first number "minvalue" and call the second number "maxvalue" and put it below the right column.

df2 = pd.DataFrame(salarylist,columns = ['salary'])


df2['minsalary'] = df2['salary'].apply(parseNumbers).apply(min)
print df2['minsalary']
# df['maxsalary'] = df['salary'].apply(parseNumbers).apply(max)
```
